### [Tycho Brahe’s Observatory](https://openstax.org/books/astronomy/pages/3-1-the-laws-of-planetary-motion)

Three years after the publication of Copernicus’ De Revolutionibus, Tycho Brahe was born to a family of Danish nobility. He developed an early interest in astronomy and, as a young man, made significant astronomical observations. Among these was a careful study of what we now know was an exploding star that flared up to great brilliance in the night sky. His growing reputation gained him the patronage of the Danish King Frederick II, and at the age of 30, Brahe was able to establish a fine astronomical observatory on the North Sea island of Hven (Figure 3.2). Brahe was the last and greatest of the pre-telescopic observers in Europe.

### [Johannes Kepler](https://openstax.org/books/astronomy/pages/3-1-the-laws-of-planetary-motion)

Johannes Kepler was born into a poor family in the German province of Württemberg and lived much of his life amid the turmoil of the Thirty Years’ War (see Figure 3.2). He attended university at Tubingen and studied for a theological career. There, he learned the principles of the Copernican system and became converted to the heliocentric hypothesis. Eventually, Kepler went to Prague to serve as an assistant to Brahe, who set him to work trying to find a satisfactory theory of planetary motion—one that was compatible with the long series of observations made at Hven. Brahe was reluctant to provide Kepler with much material at any one time for fear that Kepler would discover the secrets of the universal motion by himself, thereby robbing Brahe of some of the glory. Only after Brahe’s death in 1601 did Kepler get full possession of the priceless records. Their study occupied most of Kepler’s time for more than 20 years.

## We will use the Small Angle Approximation to calculate distances from angles

![expl](sma.jpeg)

### Target is moon - moon is 384000 KM from Earth

```python
moon_distance_wiki = 384e3 #km
print(f'Average Distance of the moon from Earth = {moon_distance_wiki:.0f} km')
```

```python
import math
measured_extent = (814.2+789.7)/2 # pixels
print(f'Measured extent of moon = {measured_extent} pixels')
arcsec_in_radian = 206265 #arc-seconds in 1 radian'
print(f'There are {arcsec_in_radian} arc-sec in 1 radian')
pixel_size_seestar_s50 = 2.9e-3 # mm
focal_length_seestar_s50 = 250 # mm
pixel_scale = arcsec_in_radian * pixel_size_seestar_s50/focal_length_seestar_s50 # small angle approximation formula
print(f'Pixel Scale of Camera/Telescope = {pixel_scale:.2f} arc-sec/pixel')
moon_pixel_angle_arcsec = pixel_scale * measured_extent
moon_dia = moon_pixel_angle_arcsec/arcsec_in_radian * moon_distance_wiki  # small angle approximation formula
print(f'Diameter of moon = {moon_dia:.0f} km')
```

```python
moon_dia_wiki = 3474.8 #km
pct_error = (moon_dia - moon_dia_wiki)/moon_dia * 100
print(f'Error in moon Diameter measurement = {pct_error:1.2f} %')
```

### Target is M81-M82 system. It is 11.8 Million Light Years away from Earth

```python
m81_m82_dist_from_earth_wiki = 11.8e6 #light years
print(f'Average Distance of the m81_m82 from Earth = {m81_m82_dist_from_earth_wiki:.2e} ly')
```

```python
import math
measured_extent = 1083.8 # pixels 
print(f'Measured extent of m81_m82 = {measured_extent} pixels')
arcsec_in_radian = 206265 #arc-seconds in 1 radian'
print(f'There are {arcsec_in_radian} arc-sec in 1 radian')
pixel_size_tv85 = 4.78e-3 # mm
focal_length_tv85 = 482*0.8 # mm with .8 focal reducer
pixel_scale = arcsec_in_radian * pixel_size_tv85/focal_length_tv85 # small angle approximation formula
print(f'Pixel Scale of Camera/Telescope = {pixel_scale:.2f} arc-sec/pixel')
m81_m82_pixel_angle_arcsec = pixel_scale * measured_extent
print(f'Measured Sky Angle between M81 & M82 = {m81_m82_pixel_angle_arcsec:.2f} arc-sec')
m81_m82_dist = m81_m82_pixel_angle_arcsec/arcsec_in_radian * m81_m82_dist_from_earth_wiki  # small angle approximation formula
print(f'Distance between m81_m82 = {m81_m82_dist:.0f} ly')
```

```python
m81_m82_dist_wiki = 150e3 #ly
pct_error = (m81_m82_dist - m81_m82_dist_wiki)/m81_m82_dist * 100
print(f'Error in m81_m82 distance measurement of m81_m82 = {pct_error:1.2f} %')
```
