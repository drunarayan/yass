| :--- | :--- | :--- |

| **Lens Focal Length** | $f$ | **22.0 mm** |

| **Pixel Size (Pitch)** | $p$ | **6.45 μm** ($0.00645$ mm) |

### **3. The Math (Small Angle Approximation)**
1.  **Linear size on sensor ($s$):** 
    $$s = \text{Measured Pixels} \times p$$
2.  **Angular size in radians ($\theta$):** 
    $$\theta \approx \frac{s}{f}$$
3.  **Actual Diameter ($d$):** 
    $$d = \theta \times D$$

### **4. Python Reference Code**
```python
# Constants
pixel_pitch = 0.00645  # mm
focal_length = 22.0    # mm
distance_to_earth = 15000.0  # km

# Input from Astrolab Stellar measurement
measured_pixels = 2925.6 

# Calculation
s = measured_pixels * pixel_pitch
theta = s / focal_length
calculated_diameter = theta * distance_to_earth

print(f"Calculated Earth Diameter: {calculated_diameter:.2f} km")
# Expected Result: ~12,865 km (vs actual 12,742 km)
